#Importando as bibliotecas

```python
import snscrape.modules.twitter as sntwitter

#Manipulação de dados
import pandas as pd
```

```python
maximo_de_tweets = 10000
```

#Lula

```python
lista_de_twitts_lula = []
```

```python
type (lista_de_twitts_lula)
```

```python
for i, tweet in enumerate(sntwitter.TwitterSearchScraper ( ['lula'] ).get_items ()):
  if i > maximo_de_tweets:
    break
  lista_de_twitts_lula.append ([tweet.date, tweet.id, tweet.content])
```

```python
dataframe_tweets_lula = pd.DataFrame(lista_de_twitts_lula, columns=['Data_Hora', 'IdTweet', 'Tweet'])
```

```python
dataframe_tweets_lula.head()
```

```python
dataframe_tweets_lula['Cadidato'] = 'Lula'
```

```python
dataframe_tweets_lula.head()
```
